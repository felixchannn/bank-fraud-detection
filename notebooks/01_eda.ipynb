```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('data/raw/banking_transactions.csv')

# Check structure
print("Shape:", df.shape)
print("\nColumn names and types:")
print(df.dtypes)
print("\nFirst few rows:")
print(df.head())

# Data quality
print("\nMissing values:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())

# Target variable (fraud)
print("\nFraud distribution:")
print(df['fraud_flag'].value_counts())
print("\nFraud percentage:")
print(df['fraud_flag'].value_counts(normalize=True) * 100)

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Correlation with fraud
print("\nCorrelation with fraud_flag:")
correlation = df.corr()['fraud_flag'].sort_values(ascending=False)
print(correlation)
```
